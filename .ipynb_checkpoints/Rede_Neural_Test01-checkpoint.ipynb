Criar o dataset

```python
def get_curva_linear(x, w, b=0, noise_scale=0): #temperatura em celcius, inclinacao da reta, e o vies(bias) = 0 valor padrao
    return w*x + b + noise_scale * np.random.randn(x.shape[0]) #equacao da reta + geracao de numeros | quanto maior o numero do noite scale maior sera o ruido
```

```python
x = np.arange(-10,30.1,0.5)
Y = get_curva_linear(x, 1.8, 32, noise_scale = 2.5)
```

Modelo

```python
#passos:
#inicializar os pesos e bias 
#FeedForward
#Calcular a perda
#BackPropagation
```

```python
#Inicializando os pesos
w = np.random.rand(1)
b = 0
```

```python
w
```

```python
def forward(inputs, w, b):
    return w*inputs + b
```

```python
def mse(Y, y):
    return (Y - y)**2
```

```python
def backpropagation(inputs, outputs, targets, w, b, lr): #entradas, valor previsto, valor real que é o alvo, ou valor que se quer chagar, learning rate, taxa de aprendizado
    dw = lr*(-2*inputs*(targets - outputs)).mean()
    db = lr*(-2*(targets-outputs)).mean()
    
    w -= dw
    b -= db
    return w, b
```

```python
def model_fit(inputs, target, w, b, epochs = 200, lr = 0.005): #ajuste do modelo
    for epoch in range(epochs): #pra cada epoca contida no range de epochs 
        outputs = forward(inputs, w, b)
        loss = np.mean(mse(target, outputs))
        w, b = backpropagation(inputs, outputs, target, w, b, lr)
        
        if(epoch + 1) % (epochs/10) == 0:
            print(f'Epoch: [{(epoch+1)}/{epochs}] Loss: [{loss:.4f}]')
    return w, b
```

```python
x = np.arange(-10, 10, 2)
Y = get_curva_linear(x, w = 1.8, b = 32, noise_scale = 2)
```

```python
#inicializacao
w = np.random.randn(1)
b = np.zeros(1)
```

```python
w, b = model_fit(x, Y, w, b, epochs = 700, lr=0.01)
print(f'w: {w[0]:.3f}, b: {b[0]:3f}')
```

```python
plt.scatter(x, Y)
plt.plot(x, get_curva_linear(x, w, b), 'r', lw = 3)
```
